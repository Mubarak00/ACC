```python
from pathlib import Path
import os

image_dir = Path("C:/Users/NET_PC/Desktop/Adrenocortical carcionoma project/JPG 2 Data")

filepaths = list(image_dir.glob(r'**/*.JPG'))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))

filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')

image_df = pd.concat([filepaths, labels], axis=1)
```

```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 28]}
df = pd.DataFrame(data)

# Create a new row as a dictionary
new_row = {'Name': 'David', 'Age': 22}

# Convert the new row dictionary into a DataFrame
new_row_df = pd.DataFrame([new_row])

# Concatenate the original DataFrame with the new row DataFrame
df = pd.concat([df, new_row_df], ignore_index=True)

print(df)
```
